13.1 Write the current date as a string to the text file today.txt.

```python
from datetime import date

# Get today's date
today = date.today()

# Convert the date to a string
today_string = str(today)

# Write the date to a text file
with open("today.txt", "w") as file:
    file.write(today_string)

print("Today's date has been written to today.txt")
```

13.2 Read the text file today.txt into the string today_string.

```python
# Open the file for reading
with open("today.txt", "r") as file:
    today_string = file.read()

print(today_string)
```

13.3 Parse the date from today_string.

```python
from datetime import datetime

# Convert the string back into a date object
parsed_date = datetime.strptime(today_string, "%Y-%m-%d")

print(parsed_date)
```

15.1 Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
import multiprocessing
import random
import time
from datetime import datetime

# Function that each process will run
def now_seconds():
    wait_time = random.random()  # Random number between 0 and 1
    time.sleep(wait_time)
    print("Current time:", datetime.now().time())

# Create three processes
if __name__ == "__main__":
    processes = []

    for i in range(3):
        process = multiprocessing.Process(target=now_seconds)
        processes.append(process)
        process.start()

    for process in processes:
        process.join()

    print("All processes have finished.")
```
